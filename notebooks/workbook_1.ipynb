```python
# Define function to reduce category count
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for category, count in categories.items():
        if count >= cutoff:
            categorical_map[category] = category
        else:
            categorical_map[category] = "Other"
    return categorical_map
```

```python
# Create the mapping with a cutoff of 400 counts
country_value_counts = df_24["country"].value_counts()
country_map = shorten_categories(country_value_counts, 400)
df_24["country"] = df_24["country"].map(country_map)

# Check the value counts after mapping
print("\nCountry Value Counts After Mapping:")
print(df_24["country"].value_counts())
```

```python
# Simplifying the employment column
def clean_employment_status(status):
    if 'Employed, full-time' in status:
        return 'Employed, full-time'
    elif 'Independent contractor, freelancer, or self-employed' in status:
        return 'Freelancer/Contractor'
    elif 'Student' in status:
        return 'Student'
    elif 'Employed, part-time' in status:
        return 'Employed, part-time'
    elif 'Retired' in status:
        return 'Retired'
    elif 'Not employed' in status:
        return 'Unemployed'
    else:
        return 'Other'

df_24['employment'] = df_24['employment'].apply(clean_employment_status)
display(df_24["employment"].value_counts())
```

```python
# Simplify the education level column with mapping function
def simplify_education_level(education_level):
    if education_level in ["Primary/elementary school", "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)"]:
        return "High School or Less"
    elif education_level in ["Some college/university study without earning a degree", "Associate degree (A.A., A.S., etc.)"]:
        return "Some College"
    elif education_level == "Bachelor’s degree (B.A., B.S., B.Eng., etc.)":
        return "Bachelor Degree"
    elif education_level == "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)":
        return "Master Degree"
    elif education_level == "Professional degree (JD, MD, Ph.D, Ed.D, etc.)":
        return "Professional/Doctoral"
    else:
        return "Other"
```

```python
# Apply the function to the education_level column
df_24['education_level'] = df_24['education_level'].apply(simplify_education_level)
print(df_24['education_level'].value_counts())
```
