```python
# Define function to reduce category count
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for category, count in categories.items():
        if count >= cutoff:
            categorical_map[category] = category
        else:
            categorical_map[category] = "Other"
    return categorical_map
```

```python
# Create the mapping with a cutoff of 400 counts
country_value_counts = df_23["country"].value_counts()
country_map = shorten_categories(country_value_counts, 400)
df_23["country"] = df_23["country"].map(country_map)

# Check the value counts after mapping
print("\nCountry Value Counts After Mapping:")
print(df_23["country"].value_counts())
```

```python
# simplify the employment column
def clean_employment_status(status):
    if isinstance(status, float) and pd.isna(status):
        return 'Other'
    elif 'Employed, full-time' in status and 'Independent contractor, freelancer, or self-employed' in status:
        return 'Freelancer/Contractor'
    elif 'Employed, full-time' in status:
        return 'Employed, full-time'
    elif 'Independent contractor, freelancer, or self-employed' in status:
        return 'Freelancer/Contractor'
    elif 'Employed, part-time' in status:
        return 'Employed, part-time'
    elif 'Retired' in status:
        return 'Retired'
    elif 'Not employed' in status:
        return 'Unemployed'
    else:
        return 'Other'

# Apply the function to the employment column
df_23['employment'] = df_23['employment'].apply(clean_employment_status)
display(df_23["employment"].value_counts())
```

```python
# Simplify the education level column with mapping function
def simplify_education_level(education_level):
    if education_level in ["Primary/elementary school", "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)"]:
        return "High School or Less"
    elif education_level in ["Some college/university study without earning a degree", "Associate degree (A.A., A.S., etc.)"]:
        return "Some College"
    elif education_level == "Bachelor’s degree (B.A., B.S., B.Eng., etc.)":
        return "Bachelor Degree"
    elif education_level == "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)":
        return "Master Degree"
    elif education_level == "Professional degree (JD, MD, Ph.D, Ed.D, etc.)":
        return "Professional/Doctoral"
    else:
        return "Other"
```

```python
# Apply the function to the education_level column
df_23['education_level'] = df_23['education_level'].apply(simplify_education_level)
print(df_23['education_level'].value_counts())
```

```python
# Grouping specific developer roles
df_23['type'] = df_23['type'].replace({
    'Developer, full-stack': 'Full Stack Developer',
    'Developer, back-end': 'Back-End Developer',
    'Developer, front-end': 'Front-End Developer',
    'Developer, desktop or enterprise applications': 'Other Developer',
    'Developer, mobile': 'Other Developer',
    'Developer, embedded applications or devices': 'Other Developer',
    'Developer, QA or test': 'Other Developer',
    'Developer, game or graphics': 'Other Developer'
})

# Grouping DevOps roles
devops_roles = ['DevOps specialist', 'Cloud infrastructure engineer', 'Engineer, site reliability']
df_23['type'] = df_23['type'].replace(devops_roles, 'DevOps')

# Grouping data roles specifically as Data Analyst, Data Scientist, Data Engineer
df_23['type'] = df_23['type'].replace({
    'Data or business analyst': 'Data Analyst',
    'Data scientist or machine learning specialist': 'Data Scientist',
    'Data engineer': 'Data Engineer'
})

# Consolidating rare roles under "Specialist"
rare_roles = ['Marketing or sales professional', 'Developer Advocate', 'Scientist', 
              'Security professional', 'Project manager', 'Product manager', 'Educator', 
              'Academic researcher', 'Research & Development role', 'Senior Executive (C-Suite, VP, etc.)']
df_23['type'] = df_23['type'].replace(rare_roles, 'Specialist')

# Handle any remaining ungrouped roles like "Other (please specify):"
df_23['type'] = df_23['type'].replace('Other (please specify):', 'Specialist')
```
